```python
import sklearn
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris.keys()
```

```python
x=iris.data
y=iris.target
```

```python
print(x)
```

```python
print(y)
```

```python
x,y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.60,test_size=0.40, random_state=47)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predicted_y=model.predict(x_test)
```

```python
predicted_y
```

```python
model.predict_proba(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
accuracy_score(y_test,predicted_y)
```

```python
a=classification_report(y_test,predicted_y)
print(a)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=0.95)
```

```python
pca.fit(x_train)
```

```python
x_train_transformed = pca.transform(x_train)
```

```python
x_train_transformed.shape
```

```python
print(x_train_transformed)
```

```python
x_train.shape
```

```python
x_test_transformed = pca.transform(x_test)
```

```python
new_model = LogisticRegression(penalty='l1', solver='liblinear')
```

```python
new_model.fit(x_train_transformed, y_train)
```

```python
new_prediction = new_model.predict(x_test_transformed)
```

```python
print (y_test)
```

```python
print(predicted_y)
```

```python
print(new_prediction)
```

```python
new_accuracy = accuracy_score(new_prediction,y_test)
new_accuracy
```

```python
new_confusion_matrix=confusion_matrix(y_test,new_prediction)
```

```python
print(new_confusion_matrix)
```

```python
cv=classification_report(y_test,new_prediction)
print(cv)
```
